```python
d = {'no': False, 'yes': True}
df['international plan'] = df['international plan'].map(d)
df.head()
```

```python
# same thinhg can be done with replace method
df = df.replace({'Voice mail plan': d})
df.head()
```

```python
#df.groupby(by=grouping_columns)[col_toshow].function()
showcol = ["total day minutes", "total day calls", "total day charge"]
df.groupby("area code")[showcol].describe(percentiles=[])
```

```python
df.pivot_table(['total day minutes', 'total day calls', 'total day charge'], ['area code'], aggfunc = 'mean')
```

```python
total_calls = df['total day calls']+ df['total eve calls'] + \
df['total night calls'] + df['total intl calls']
df.insert(loc=len(df.columns), column='total calls', value=total_calls)
```
